```python
import orgelpredigt_analysis as oa
```

```python
# create list of sentences for sermon
id = "E000052"
sermon = oa.Sermon(id)
```

```python
words_and_types_no_bible = [[x, y, z] for x, y, z in zip(sermon.words, sermon.word_types, sermon.reference) if y != " bibel" and isinstance(x, str)]
only_words_no_bible = [x for x, y in zip(sermon.words, sermon.word_types) if y != " bibel" and isinstance(x, str)]
```

```python
def chunk(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i+n]
```

```python
sermon_sents = list(chunk(only_words_no_bible, 10))
```

```python
sermon_vec = Word2Vec(sermon_sents)
```

```python
def remove_sw(lst, sw):
    return [w.lower() for w in lst if w not in sw]
```

```python
sermon_words = [remove_sw(s, sw) for s in sermon_sents]
sermon_vec = Word2Vec(sermon_words)
```
